```python
import pandas as pd
from utils import Article
```

```python
data=pd.read_csv('./dataset/news.tsv',header=None,sep='\t')
```

```python
url = 'https://assets.msn.com/labs/mind/AAGH0ET.html'
```

```python
data.columns=['News ID',
"Category",
"SubCategory",
"Title",
"Abstract",
"URL",
"Title Entities",
"Abstract Entities "]

```

```python
data.dropna(inplace=True)
```

```python
urls = data['URL'].to_list()
```

```python
article_dict={}
```

```python
i=0
for url in urls:
    article_dict[url] = Article.scraper(url)
    i+=1
    if i % 250 == 0:
        print(i)
```

```python
article_df=pd.DataFrame(list(article_dict.items()), columns=['URL', 'Article'])
```
